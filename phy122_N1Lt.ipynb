```python
x_t1 = zeros((len(timeline), 2)) ## initialize position-trajectory (position vectors for all times)
v_t1 = zeros((len(timeline), 2)) ## initialize velocity-trajectory (velocity vectors for all times)

x_t1[0] = x_i1 ## initialize the 0th position-trajectory entry to the initial-position
for t in range(1,len(timeline)):
    x_t1[t] = x_t1[t-1] + v_t1[t-1] * dt ## numerically integrate the velocity
    v_t1[t] = v_t1[t-1] ## set new velocity = old velocity (because constant velocity)
```
