#   <font color='Red'> **Import Necessary Modules** </font>

```python
import numpy as np
from numpy.linalg import norm
from numpy import array, sqrt, zeros, arange
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import textwrap

π = np.pi

def d(X):
    return np.gradient(X, axis=0)
```

### <font color='Red'> Before we start lets define the times in which all our experiments will take place. </font>

### <font color='Red'> Define a time-increment (Δt), initial (t0), and a final (tf) time for your experiment. And with it a time_array. </font>

```python
dt = 0.01 ### s(detector)
t0 = 0
tf = 10
timeline = arange(t0, tf, dt)
```

#   <font color='Red'> **Experiment 1** </font>

### <font color='Red'> To begin lets start with by defining an initial position. </font>

```python
x_i1 = array([1., 0., 0.]) ## m : initial position vector/array
```

### <font color='Red'> PRINT the distance of the previously defined position vector (give units!): </font>

```python
distance1 = norm(x_i1)

print("the initial distance is %4.3f meters "  % distance1)
```

### <font color='Red'> Define a for-loop to Velocity-Verlet integrate the motion to obtain a x_t. </font>

```python
x_t1 = zeros((len(timeline), 3)) ## initialize position-trajectory (position vectors for all times)
v_t1 = zeros((len(timeline), 3)) ## initialize velocity-trajectory (velocity vectors for all times)

x_t1[0] = x_i1 ## initialize the 0th position-trajectory entry to the initial-position
for t in range(1,len(timeline)):
    x_t1[t] = x_t1[t-1] + v_t1[t-1] * dt ## numerically integrate the velocity
    v_t1[t] = v_t1[t-1] ## set new velocity = old velocity (because constant velocity)
```

### <font color='Red'> Take a time-derivative of x_t to obtain dxdt_t. </font>

```python
dxdt_t1 = d(x_t1)/dt ## take numerical-derivative of the aforementioned position-trajectory
```

### <font color='Red'> Make 2 plots of with: x_t and \{v_t, dxdt_t\} in matplotlib, both as a function of time. Make sure to include all LABELS!. </font>

```python
fig1, (ax11, ax21) = plt.subplots(1, 2)

fig1.text(1, 1, "Experiment 1", ha='center', fontsize='x-large')
fig1.subplots_adjust(hspace=0.01, wspace=0.25, right=2, top=0.93, bottom=0.17)
caption1 = "Experiment 1 begins with an initial position of 1 m and an initial velocity of 0 m/s. In the absence of any acceleration, the initial conditions remain constant for the entire trial."\
            " Velocity-Verlet result (v(t)) is indistinguishable from direct derivative of the position function (dx(t)/dt)."
caption_wrapped1 = textwrap.fill(caption1, width=150)
fig1.text(1, 0, caption_wrapped1, ha='center')

ax11.scatter(timeline, x_t1[:, 0], label="x(t)")
ax11.set_title("Position")
ax11.set_xlim(0, 10)
ax11.set_ylim(-5, 5)
ax11.legend()
ax11.set(xlabel="time (s)", ylabel="position (m)")

ax21.scatter(timeline, v_t1[:, 0], label="v(t)")
ax21.scatter(timeline, dxdt_t1[:, 0], label="dx(t)/dt")
ax21.set_title("Velocity")
ax21.set_xlim(0, 10)
ax21.set_ylim(-5, 5)
ax21.legend()
ax21.set(xlabel="time (s)", ylabel="velocity (m/s)")

fig1.show()
```

#   <font color='Red'> **Experiment 3** </font>

###  <font color='Red'> To begin lets start with by defining an initial position, velocity, and acceleration </font>

```python
x_i3 = array([0., 0., 0.]) ## m : initial position vector/array
v_i3 = array([0., 0., 0.]) ## m/s : initial velocity vector/array
a_i3 = array([1., 0., 0.]) ## m/s/s : initial acceleration vector/array
```

###  <font color='Red'> PRINT out the distance, speed, and magnitude of the acceleration of the previously defined vectors: </font>

```python
distance3 = norm(x_i3)
speed3 = norm(v_i3)
acceleration3 = norm(a_i3)

print("the initial distance is %4.3f meters "  % distance3)
print("the initial speed is %4.3f meters/second "  % speed3)
print("the initial acceleration is %4.3f meters/second/second "  % acceleration3)
```

###  <font color='Red'> Define a for-loop to Velocity-Verlet integrate the motion to obtain a x_t, v_t, and a_t. </font>

```python
x_t3 = zeros((len(timeline), 3)) ## initialize position-trajectory (position vectors for all times)
v_t3 = zeros((len(timeline), 3)) ## initialize velocity-trajectory (velocity vectors for all times)
a_t3 = zeros((len(timeline), 3)) ## initialize acceleration-trajectory (acceleration vectors for all times)

x_t3[0] = x_i3 ## initialize the 0th position-trajectory entry to the initial-position
v_t3[0] = v_i3 ## initialize the 0th velocity-trajectory entry to the initial-velocity
a_t3[0] = a_i3 ## initialize the 0th acceleration-trajectory entry to the initial-acceleration
for t in range(1,len(x_t3)): ## iteratively run algorithm for all entries (for all time, excluding time 0)
  x_t3[t] = x_t3[t-1] + v_t3[t-1] * dt + 0.5 * a_t3[t-1] * dt**2 ## Velocity-Verlet Step 1 : find new position from old data
  a_t3[t] = a_t3[t-1] ## Velocity-Verlet Step 2 : find new acceleraiton from old acceleraiton
  v_t3[t] = v_t3[t-1] + 0.5 * ( a_t3[t-1] + a_t3[t] ) * dt ## Velocity-Verlet Step 3 : find new velociy from old data & new accleration
```

###  <font color='Red'> Take a time-derivative of x_t to obtain dxdt_t. </font>

###  <font color='Red'> Take a time-derivative of v_t to obtain dvdt_t. </font>

###  <font color='Red'> Take a time-derivative of a_t to obtain dadt_t. </font>

```python
dxdt_t3 = d(x_t3)/dt ## take numerical-derivative of the aforementioned position-trajectory
dvdt_t3 = d(v_t3)/dt ## take numerical-derivative of the aforementioned velocity-trajectory
dadt_t3 = d(a_t3)/dt ## take numerical-derivative of the aforementioned acceleration-trajectory
```

###  <font color='Red'> Make 4 plots of with: x_t, \{v_t, dxdt_t\}, \{ a_t, dvdt_t\}, and dadt_t in matplotlib, all as a function of time. Make sure to include all LABELS!. </font>

```python
fig3, (ax13, ax23, ax33, ax43) = plt.subplots(1, 4)

fig3.text(1, 1, "Experiment 3", ha='center', fontsize='x-large')
fig3.subplots_adjust(hspace=0.01, wspace=0.25, right=2, top=0.93, bottom=0.2)
caption3 = "Experiment 3 begins with an initial position of 0 m, an initial velocity of 0 m/s, and an initial acceleration of 1 m/s\u00b2."\
            " The rate of change of acceleration (jerk) is 0 m/s\u00b3, so the initial acceleration remains constant for the entire trial."\
            " The positive, constant, acceleration causes velocity to increase linearly with time, which in turn causes position to increase exponentially with time."\
            " Velocity-Verlet results v(t), a(t) are indistinguishable from their corresponding direct derivatives dx(t)/dt, dv(t)/dt."
caption_wrapped3 = textwrap.fill(caption3, width=175)
fig3.text(1, 0, caption_wrapped3, ha='center')

ax13.scatter(timeline, x_t3[:, 0], label="x(t)")
ax13.set_title("Position")
ax13.legend()
ax13.set(xlabel="time (s)", ylabel="position (m)")

ax23.scatter(timeline, v_t3[:, 0], label="v(t)")
ax23.scatter(timeline, dxdt_t3[:, 0], label="dx(t)/dt")
ax23.set_title("Velocity")
ax23.legend()
ax23.set(xlabel="time (s)", ylabel="velocity (m/s)")

ax33.scatter(timeline, a_t3[:, 0], label="a(t)")
ax33.scatter(timeline, dvdt_t3[:, 0], label="dv(t)/dt")
ax33.set_xlim([0, 10])
ax33.set_ylim([-5, 5])
ax33.set_title("Acceleration")
ax33.legend()
ax33.set(xlabel="time (s)", ylabel="acceleration (m/s\u00b2)")

ax43.scatter(timeline, dadt_t3[:, 0], label="da(t)/dt")
ax43.set_xlim([0, 10])
ax43.set_ylim([-5, 5])
ax43.set_title("Jerk")
ax43.set(xlabel="time (s)", ylabel="jerk (m/s\u00b3)")
ax43.legend()

fig3.show()
```

“For cases of constant acceleration we may generalize the integrator to (also known as the Velocity-Verlet Integrator)

1)	x[t] = x[t−1] + (v[t−1])dt + 0.5(a[t−1])^2
