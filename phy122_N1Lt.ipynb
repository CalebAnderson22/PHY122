#   <font color='Red'> **Import Necessary Modules** </font>

#   <font color='Red'> **Experiment 3** </font>

###  <font color='Red'> To begin lets start with by defining an initial position, velocity, and acceleration </font>

```python
x_i3 = array([1., 1.]) ## m : initial position vector/array
v_i3 = array([1., 1.]) ## m/s : initial velocity vector/array
a_i3 = array([1., 1.]) ## m/s/s : initial acceleration vector/array
```

###  <font color='Red'> PRINT out the distance, speed, and magnitude of the acceleration of the previously defined vectors: </font>

```python
distance = norm(x_i3)
speed = norm(v_i3)
acceleration = norm(a_i3)

print("the initial distance is %4.3f meters "  % distance)
print("the initial speed is %4.3f meters/second "  % speed)
print("the initial acceleration is %4.3f meters/second\u00b2 "  % acceleration)
```

###  <font color='Red'> Define a for-loop to Velocity-Verlet integrate the motion to obtain a x_t, v_t, and a_t. </font>

```python
x_t3 = zeros((len(timeline), 2)) ## initialize position-trajectory (position vectors for all times)
v_t3 = zeros((len(timeline), 2)) ## initialize velocity-trajectory (velocity vectors for all times)
a_t3 = zeros((len(timeline), 2)) ## initialize acceleration-trajectory (acceleration vectors for all times)

x_t3[0] = x_i3 ## initialize the 0th position-trajectory entry to the initial-position
v_t3[0] = v_i3 ## initialize the 0th velocity-trajectory entry to the initial-velocity
a_t3[0] = a_i3 ## initialize the 0th acceleration-trajectory entry to the initial-acceleration
for t in range(1,len(x_t3)): ## iteratively run algorithm for all entries (for all time, excluding time 0)
  x_t3[t] = x_t3[t-1] + v_t3[t-1] * dt + 0.5 * a_t3[t-1] * dt**dt ## Velocity-Verlet Step 1 : find new position from old data
  a_t3[t] = a_t3[t-1] ## Velocity-Verlet Step 2 : find new acceleraiton from old acceleraiton
  v_t3[t] = v_t3[t-1] + 0.5 * ( a_t3[t-1] + a_t3[t] ) * dt ## Velocity-Verlet Step 3 : find new velociy from old data & new accleration
```

###  <font color='Red'> Take a time-derivative of x_t to obtain dxdt_t. </font>

###  <font color='Red'> Take a time-derivative of v_t to obtain dvdt_t. </font>

###  <font color='Red'> Take a time-derivative of a_t to obtain dadt_t. </font>

```python
dxdt_t3 = d(x_t3)/dt ## take numerical-derivative of the aforementioned position-trajectory
dvdt_t3 = d(v_t3)/dt ## take numerical-derivative of the aforementioned velocity-trajectory
dadt_t3 = d(a_t3)/dt ## take numerical-derivative of the aforementioned acceleration-trajectory
```

###  <font color='Red'> Make 4 plots of with: x_t, \{v_t, dxdt_t\}, \{ a_t, dvdt_t\}, and dadt_t in matplotlib, all as a function of time. Make sure to include all LABELS!. </font>

```python
fig3, (ax13, ax23, ax33, ax43) = plt.subplots(1, 4)
ax13.axis('off')
ax23.axis('off')
ax33.axis('off')
ax43.axis('off')
ax13 = fig3.add_subplot(1, 4, 1, projection='3d')  # Create a 3D subplot
ax23 = fig3.add_subplot(1, 4, 2, projection='3d')  # Create a 3D subplot
ax33 = fig3.add_subplot(1, 4, 3, projection='3d')  # Create a 3D subplot
ax43 = fig3.add_subplot(1, 4, 4, projection='3d')  # Create a 3D subplot

fig3.text(1, 1, "Experiment 3", ha='center', fontsize='x-large')
fig3.subplots_adjust(hspace=0.01, wspace=0.25, right=2, top=0.95, bottom=0.15)
caption3 = "caption3"
caption_wrapped3 = textwrap.fill(caption3, width=40)
fig3.text(1, 0, caption_wrapped3, ha='center')

ax13.scatter(x_t3[:, 0], x_t3[:, 1], timeline, label="x(t)")
ax13.set_title("Distance")
ax13.legend()
ax13.set(xlabel="time (s)", ylabel="Distnce from Origin (m)")

ax23.scatter(v_t3[:, 0], v_t3[:, 1], timeline, label="v(t)")
ax23.scatter(dxdt_t3[:, 0], dxdt_t3[:, 1], timeline, label="dx(t)/dt")
ax23.set_title("Speed")
ax23.legend()
ax23.set(xlabel="time (s)", ylabel="Speed (m/s)")

ax33.scatter(a_t3[:, 0], a_t3[:, 1], timeline, label="a(t)")
ax33.scatter(dvdt_t3[:, 0], dvdt_t3[:, 1], timeline, label="dv(t)/dt")
ax33.set_ylim([0, 4])
ax33.set_title("Acceleration")
ax33.legend()
ax33.set(xlabel="time (s)", ylabel="Acceleration (m/s$^2$)")

ax43.scatter(dadt_t3[:, 0], dadt_t3[:, 1], label="da(t)/dt")
ax43.set_ylim([0, 4])
ax43.set_title("Jerk")
ax43.set(xlabel="time (s)", ylabel="Jerk (m/s$^3$)")
ax43.legend()

fig3.show()
```
